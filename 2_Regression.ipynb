```python
import pandas as pd                   # Library for data manipulation and analysis
import numpy as np                    # For reshaping arrays (numpy)
from sklearn.model_selection import train_test_split  # For splitting data into training and testing sets
from sklearn.linear_model import LinearRegression  # For creating and using a linear regression model
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score  # Evaluation metrics
import matplotlib.pyplot as plt        # For creating visualizations

```

```python
df = pd.read_csv('/temp.csv')
```

```python
# dependent variable is year
x = df['YEAR']

# independant variable is
y =df['ANNUAL']
```

```python
plt.title("temp plot of india year vs jan")
plt.scatter(x,y)
plt.xlabel("year")
plt.ylabel("jan")


```

```python
x.shape
```

```python
x=x.values
```

```python

# scikit-leran libaray requirment is of 2d array
x=x.reshape(117,1)
```

```python
x.shape
```

```python
type(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=0.25,random_state=0)

print(f"x Training dataset: {x_train.shape}")
print(f"y Training dataset: {y_train.shape}")
print(f"x test dataset: {x_test.shape}")
print(f"y test dataset: {y_test.shape}")
```

```python
# create object of class "Linearregression"
regressor = LinearRegression()
```

```python
regressor.fit(x_train,y_train)
```

```python
# y= mx +c
#  m= slope of regression line called as coefficent calculated by model

regressor.coef_
```

```python
#  c = y intercept of regression line calculated by model
regressor.intercept_
```

```python
y_predicted =regressor.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
print(f"MSE:  {mean_squared_error(y_test,y_predicted)}")
print(f"MAE:  {mean_absolute_error(y_test,y_predicted)}")
print(f"R-Sqaure :  {r2_score(y_test,y_predicted)}")
```

```python
np.mean(abs(y_test- y_predicted))
```

```python
np.mean((y_test- y_predicted)**2)
```

```python
regressor.score(x_test,y_test)
```

```python
plt.scatter(x_train, y_train, color='blue' )
plt.plot(x_test, y_predicted, color='red', linewidth=3 )

plt.title("Temperature vs Year")
plt.xlabel("Year")
plt.ylabel("Temperature")
plt.show()
```
